```python
#print(X)
from sklearn.datasets import make_moons
from matplotlib import pyplot as plt 
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
import numpy as np
np.random.seed(50)
X, y = make_moons(n_samples=500, noise=0.1)
```

```python
plt.scatter(X[:,0], X[:,1] ,c=y )
```

```python
def get_random(k):
    centroids={}
    while True:
        ri=np.random.choice(range(len(X)), k)
        print(ri)
        if len(ri)==k:
            for i in range(k):
                centroids[i]=X[ri[i]]
            break

    return centroids
```

```python
k = 8
c=get_random(k)
print(c)
c=list(c.values())
print(c)
c1=c[0]
c2=c[1]
c3=c[2]
c4=c[3]
c5=c[4]
c6=c[5]
c7=c[6]
c8=c[7]
```

```python
def euclidianDistance(arr,c):
    return np.sqrt( ((arr[0]-c[0])*(arr[0]-c[0]))+((arr[1]-c[1])*(arr[1]-c[1])) )
```

```python
def grouping(X):
    g1 = []
    g2 = []
    g3 = []
    g4 = []
    g5 = []
    g6 = []
    g7 = []
    g8 = []
   
    
    sz = X.size
    sz/=2
    sz = int (sz)
    print(sz)
    for i in range(sz):
        data = X[i]
        mn1 = euclidianDistance(data,c1)
        mn2 = euclidianDistance(data,c2)
        mn3 = euclidianDistance(data,c3)
        mn4 = euclidianDistance(data,c4)
        mn5 = euclidianDistance(data,c5)
        mn6 = euclidianDistance(data,c6)
        mn7 = euclidianDistance(data,c7)
        mn8 = euclidianDistance(data,c8)
       
        if mn1<mn2 and mn1<mn3 and mn1<mn4 and mn1<mn5 and mn1<mn6 and mn1<mn7 and mn1<mn8:
            g1.append(data)
        elif mn2<mn1 and mn2<mn3 and mn2<mn4 and mn2<mn5 and mn2<mn6 and mn2<mn7 and mn2<mn8:
            g2.append(data)
        elif mn3<mn1 and mn3<mn2 and mn3<mn4 and mn3<mn5 and mn3<mn6 and mn3<mn7 and mn3<mn8:
            g3.append(data)
        elif mn4<mn1 and mn4<mn2 and mn4<mn3 and mn4<mn5 and mn4<mn6 and mn4<mn7 and mn4<mn8:
            g4.append(data)
        elif mn5<mn1 and mn5<mn2 and mn5<mn3 and mn5<mn4 and mn5<mn6 and mn5<mn7 and mn5<mn8:
            g5.append(data)
        elif mn6<mn1 and mn6<mn2 and mn6<mn3 and mn6<mn4 and mn6<mn5 and mn6<mn7 and mn6<mn8:
            g6.append(data)
        elif mn7<mn1 and mn7<mn2 and mn7<mn3 and mn7<mn4 and mn7<mn6 and mn7<mn5 and mn7<mn8:
            g7.append(data)
        else:
            g8.append(data)
    return g1,g2,g3,g4,g5,g6,g7,g8
```

```python
def average(grp):
    GRP = np.array(grp)
    sz = GRP.size
    sz/=2
    sz = int (sz)
    left=0
    right=0
    print(sz)
    for i in range(sz):
        t = GRP[i]
        left+=t[0]
        right+=t[1]
    c = [left/sz,right/sz]
    return c
```

```python
g1,g2,g3,g4,g5,g6,g7,g8= grouping(X)
c1 = average(g1)
c2 = average(g2)
c3 = average(g3)
c4 = average(g4)
c5 = average(g5)
c6 = average(g6)
c7 = average(g7)
c8 = average(g8)


while(1):
    g1,g2,g3,g4,g5,g6,g7,g8 = grouping(X)
    c1 = average(g1)
    c2 = average(g2)
    c3 = average(g3)
    c4 = average(g4)
    c5 = average(g5)
    c6 = average(g6)
    c7 = average(g7)
    c8 = average(g8)
    g1,g2,g3,g4,g5,g6,g7,g8 = grouping(X)
    c11 = average(g1)
    c22= average(g2)
    c33 = average(g3)
    c44 = average(g4)
    c55 = average(g5)
    c66 = average(g6)
    c77 = average(g7)
    c88 = average(g8)
    if c1==c11 and c2==c22 and c3==c33 and c4==c44 and c5==c55 and c6==c66 and c7==c77 and c8==c88:
        break
```
